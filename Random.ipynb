```python
#load dependencies:
import pandas as pd
import numpy as np
import random

# File to Load:
file_to_load = "Resources/dnd_jobs.xlsx"

# Read Purchasing File and store into Pandas data frame:
jobs = pd.read_excel(file_to_load)

#Display preview of data frame:
jobs
```

```python
ag=len(jobs["Ag"].unique())-2
num_races=len(races["RACE"].unique())-1
ag
```

```python
num_npcs=1
while num_npcs>0:
    num_npcs=int(input("How many npcs? "))
    listicle=list(range(0,num_npcs))
    npcs=[]
    job_classes=list(jobs.columns.values)
    genders=["male", "female", "male", "female","nonbinary"]

    for l in listicle:
        gender_num=random.randint(0,4)
        gender=genders[gender_num]

        race_num=random.randint(0, num_races)
        rand_race=races.iloc[race_num,0]

        low=races.iloc[race_num,1]
        high=races.iloc[race_num, 2]
        age_range=[low,high]
        age=random.randint(0, high)

        if age<(low/2):
            age_des='child'
            rand_job= ''
            job_class=''
        else:
            if age<low:
                age_des='teen'
            if age<np.mean(age_range):
                age_des='adult'
            else:
                age_des='old'
            job_class_num=random.randint(0, len(job_classes)-1)
            job_class=job_classes[job_class_num]
            num_jobs=len(jobs[job_class].unique())-2
            job_num=random.randint(0, num_jobs)
            rand_job=jobs.iloc[job_num, job_class_num]

        npcs.append(f'{rand_race} {rand_job} ({job_class}), {age_des} {gender} ({age} years old)')
    for npc in npcs:
        print(npc)
```
